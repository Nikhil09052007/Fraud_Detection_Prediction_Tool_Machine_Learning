## Preparing tings up for implementation of and Traing the model from Machine Learning techniques. 

```python
from sklearn.model_selection import train_test_split # this package will help us to split our dataframe in various sets for training.
from sklearn.preprocessing import StandardScaler   # For scaling the data 
from sklearn.linear_model import LogisticRegression  # Modeling site
from sklearn.metrics import classification_report, confusion_matrix # model evaluation
from sklearn.pipeline import Pipeline # Traning and doing transformation opertion together
from sklearn.compose import ColumnTransformer # 
from sklearn.preprocessing import OneHotEncoder
```

```python
df.head()
```

```python
# Removing some unwanted table frames or data frames.
df_model = df.drop(["nameOrig","nameDest","isFlaggedFraud"],axis =1)
```

```python
df_model.head()  # Unwanted columns are removed now.
```

```python
# Setting up the data type according to the distribution
categorical = ["type"] #type column is categorical
numeric = ["amount","oldbalanceOrg","newbalanceOrig","oldbalanceDest","newbalanceDest"]
```

```python
y = df_model["isFraud"]
X = df_model.drop("isFraud",axis = 1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3,stratify=y)  # here we are training 70% of our data and testing 30% of it.
```

```python
# Doing the prprocessing
preprocessor = ColumnTransformer(
    transformers = [
        ("num",StandardScaler(),numeric), # For numerical type data StandardScalar
        ("cat",OneHotEncoder(drop="first"),categorical)
],
remainder = "drop"

)
```

```python
# creating a model Pipeline

pipeline = Pipeline([
    ("prep",preprocessor), # FOR preparation
    ("clf",LogisticRegression(class_weight="balanced", max_iter = 1000))    
]) 
```

```python
# Training the model with a preprocessing
pipeline.fit(X_train, y_train)
```

Here our model is ready. And now we can do the predictions of the model using this model

```python
# Testing using predict 
y_pred = pipeline.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
pipeline.score(X_test,y_test) * 100  

# This is our final percent of accuracy of the model and it do can be precised more.
```
